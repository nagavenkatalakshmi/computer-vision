```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import pylab
# %matplotlib inline
#from google.colab.patches import cv2_imshow
```

```python
flower = cv2.imread("sunflower.jpg")
```

```python
plt.imshow(flower)
```

```python
flower_gray = cv2.cvtColor(flower, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(flower_gray,cmap='gray')
```

Convolution kernel

```python
kernel = np.ones((5,5),np.float32)/25
```

```python
flower2d = cv2.filter2D(flower_gray,-1,kernel)
plt.imshow(flower2d,cmap='gray')
```

```python
bilfilt_flower = cv2.bilateralFilter(flower_gray,9,50,50)
plt.imshow(bilfilt_flower,cmap='gray')
```

```python
sobelxy = cv2.Sobel(flower_gray,cv2.CV_64F, dx=1, dy=0, ksize = 5)
```

```python
plt.imshow(sobelxy,cmap='gray')
```

```python
flower_blur = cv2.GaussianBlur(flower_gray, (3,3), 0)
plt.imshow(flower_blur,cmap='gray')
```

```python
sobelxy = cv2.Sobel(flower_blur,cv2.CV_64F, dx=1, dy=0, ksize = 5)
plt.imshow(sobelxy,cmap='gray')
```

```python
cannyedge = cv2.Canny(flower_blur, threshold1 = 100, threshold2 = 100)
```

```python
plt.imshow(cannyedge,cmap='gray')
```
